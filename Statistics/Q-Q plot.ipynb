# Q-Q plot

We have a dataset X, suppose we want to know if this dataset is gausssian distributed. For that we have various statistical tests that we can perform on the dataset to find out if it follows gaussian distribution.
these tests are 
1. KS test
2. AD(Anderson Darling) test-- very string test for checking normality of dataset.
3. Q-Q test(Graphical test)

To know how QQ plot works:
    Suppose u have dataset X having 500 datapoints, such that X[x1,x2-----,x500]
    step1: Sort Xi's in ascending order and compute percentiles 
        such that x5`,x10`------,x500`
    
    step2: simlarly create a dataset Y such that it belongs to gaussian distributions and repeat process in step1 for them.
    
    step3: Plot QQ plot keeping dataset Y in x axis known as theoritical axis and dataset X in y axis
        
    Now if the plot shows a linear relationship then we can conclude that dataset X is Normally distributed. If the relation is not linear then the dataset X doesnot belong to normal distribution.
    
    As a matter of fact we can check any distribution by changing the theoritical axis data if relation is linear then X and Y belongs to similar distribution else they are from different distribution.

```python
import numpy as np
import scipy.stats as stats
import pylab
import matplotlib.pyplot as plt
```

```python
std_normal = np.random.normal(loc=0, scale=1,size=500)
for i in np.arange(0,101):
    print(i,np.percentile(std_normal,i))  #this can be used as my Theoritical axis 
```

```python
measurements = np.random.normal(loc=30,scale=2,size=100) # this is my dataset that I wish to check

# Inspite of doing all this we can directly use the scipy.stats function to initialise 
# it with Normal distribution and plot the QQ plot

stats.probplot(x=measurements,dist='norm',plot=plt)
plt.show()
```

Here as we can see that the dataset we have is almost fall on the straight line i.e have linear relationship. A little of deviation is because of the datasize

```python
# same code as above with different datasize
measurements = np.random.normal(loc=30,scale=2,size=1000) # this is my dataset that I wish to check

stats.probplot(x=measurements,dist='norm',plot=plt)
plt.show()
```

as we increased the datasize from 100 to 1000 the data difference is visible that almost all my datapoints fall on the staright line.


```python
# lets increase the datasize further
measurements = np.random.normal(loc=30,scale=2,size=50000) # this is my dataset that I wish to check
stats.probplot(x=measurements,dist='norm',plot=plt)
plt.show()
```

As we increased my datasize further all my points are now on the straight line. This shows that my dataset measurement is normally distributed.

Limitation: If dataset size is small then it is difficult to interpret our QQ plot

```python
# Now lets suppose my measurements dataset belongs to Uniform ditribution
measurements = np.random.uniform(high=1,low=0,size=1000)

stats.probplot(x=measurements,dist='norm',plot=plt)
plt.show()
```

It is clearly visible that the datapoints doesnot lie on the staight line and hence be concluded that dataset on X and Y axis are not similar in this case. They does not belong to distribution of same family.
